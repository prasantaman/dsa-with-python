```python
from abc import ABC,abstractmethod
class BankApp(ABC):
    def database(self):
        print("connected to database")

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def display(self):
        pass
```

```python
class MobileApp(BankApp):
    def mobile_login(self):
        print("login into mobile")
    def security(self):
        print('mobile security') 
    def display(self):
        print('display security')
```

```python
mob = MobileApp()
```

```python
mob.database()
```

```python
mob.security()
```

```python
mob.display()
```
