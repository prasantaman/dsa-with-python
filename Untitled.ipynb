# Membership Operators

```python
fruits = ["apple", "banana", "cherry"]

print("apple" in fruits)       # True
print("mango" not in fruits)   # True
print("banana" not in fruits)  # False
```

# Identity Operators

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)       # True  (same memory)
print(a is c)       # False (different memory)
print(a == c)       # True  (values same)
print(a is not c)   # True
```

# Frozenset(data type)

```python
# normal set (mutable)
s = {1, 2, 3}
s.add(4)        # works
print(s)        # {1, 2, 3, 4}

# frozen set (immutable)
fs = frozenset([1, 2, 3])
# fs.add(4)  ❌ Error (not allowed)

print(fs)       # frozenset({1, 2, 3})
```

```python
A = frozenset([1, 2, 3, 4])
B = frozenset([3, 4, 5, 6])

print(A | B)   # Union → frozenset({1, 2, 3, 4, 5, 6})
print(A & B)   # Intersection → frozenset({3, 4})
print(A - B)   # Difference → frozenset({1, 2})
```

# Binary (datatype)

## Bytes

```python
data = b"Hello"
print(data)         # b'Hello'
print(data[0])      # 72  (ASCII of 'H')

# data[0] = 65 ❌ (Error — immutable)
```

# Bytearray

```python
data = bytearray([65, 66, 67])   # A, B, C
print(data)          # bytearray(b'ABC')

data[0] = 68         # change A (65) to D (68)
print(data)          # bytearray(b'DBC')
```

# Memoryview

```python
data = bytearray(b"Python")# normal string, list, tuple, ya integer par memoryview() nahi bana sakte
mv = memoryview(data)

print(mv[0])     # 80  (ASCII of 'P')

mv[0] = 112      # change to lowercase 'p'
print(data)      # bytearray(b'python')
```
