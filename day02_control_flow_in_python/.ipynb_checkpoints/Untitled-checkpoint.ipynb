```python
def greet(name,age):
    print(f"hello {name}, you are a great person {age}")
greet(name = "Prasant", age = 23)  
```

```python
def add_numbers(*args):
    total = sum(args)
    print(total)
add_numbers(5,10)    
```

```python
def user_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")
user_info(name="Prasant", classs = "10th", subject = "computer science")        
```

```python
def user_info(**kwarg): 
    for key in kwarg.items():
        print(key)
user_info(name1="Prasant",name2 = "Pragati")        
```

```python
def decorator_function(original_function):
    def wrapper_function():
        print("Before the function call")
        original_function()
        print("After the function call")
    return wrapper_function
@decorator_function 
def say_hello():
    print("Hello, Prasant!")
    
say_hello()    
```
