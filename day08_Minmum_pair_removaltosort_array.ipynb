```python
def minimumPairRemoval(nums):
    ops = 0
    i = 0

    while i < len(nums) - 1:
        if nums[i] <= nums[i + 1]:
            i += 1
        else:
            # merge violating pair
            nums[i] = nums[i] + nums[i + 1]
            nums.pop(i + 1)
            ops += 1

            # step back to re-check
            if i > 0:
                i -= 1

    return ops
```

```python
class Solution(object):
    def minimumPairRemoval(self, nums):
        ops = 0
        i = 0

        while i < len(nums) - 1:
            if nums[i] <= nums[i + 1]:
                i += 1
            else:
                nums[i] += nums[i + 1]
                nums.pop(i + 1)
                ops += 1

                if i > 0:
                    i -= 1

        return ops
```
