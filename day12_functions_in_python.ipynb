```python
def is_even(num):
    """ even number calculation"""
    if type(num) == int:
        if num % 2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return f"Pagal hai kya {type(num)} likh raha hai"
```

```python
for i in range(0,11):
    x = is_even(i)
    print(x)
```

# types of arguments
- default arguments 
- positional arguments
- keyword arguments 

```python
# default argument
def power(a,b):
    return a**b
```

```python
power(2,3)
```

```python
power(2)
```

```python
# for this kind of error we use default argument
def power(a = 1,b=3):
    print(a**b)
power(2)    
```

```python
# positional argument 
power(2,3)
```

```python
# keyword argument 
power(b=3,a=2)
```

# *args and **kwargs 

```python
# special python keyword that are used to pass the variable length of arguments to a function
```

```python
# *args allows us to pass a variable number of non_keyword arguments to a function 
def multiply(*args):# here istead of args we can use any keyword like *puppy etc 
    product = 1
    for i in args:
        product = product * i
    print(args)    
    return product    
```

```python
multiply(2,3,4,5,6)
```

```python
# *kwargs allows us to pass any number of keyword arguments.
# keyword arguments mean that they contain a key-value pair, like a python dictonary
def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,"->", value)
```

```python
display(india = 'delhi', pak = "colony")
```
