# adding items to a list 

```python
# append
l.append(True) # add items at the end
print(l)
```

```python
# extend(used to add multiple items at the end)
l.extend([6,7,8])
print(l)
```

```python
l.append([4,3,3])
print(l)
```

```python
l.extend('hello')
print(l)
```

```python
# insert
l = [1,2,3]
l.insert(1,True)
print(l)
```

# Editing items in a list 

```python
l = [1,2,3]
l[-1]= 100
print(l)
# editing with indexing
l[1:2] = [300,400]
print(l)
```

 # Deleting items from a list 

```python
# del
l = [1,2,3,4]
print(l)
del l
print(l)
```

```python
# del with indexing
l = [1,2,3,4]
del l[-1]
print(l)
```

```python
# del with slicing 
l = [1,2,3,4]
del l[1:4]
print(l)
```

```python
# remove 
l = [1,2,3,4]
l.remove(3)
print(l)
```

```python
# pop 
l = [1,2,3,4]
l.pop() # defauld index [-1]
print(l)
```

```python
# pop withot indexing 
l = [1,2,3,4]
l.pop(2)
print(l)
```

```python
# clear(to make empty list) 
l = [1,2,3,4]
l.clear()
print(l)
```

# Operators in list 

```python
# Arithmetic 
l = [1,2,3,4]
l1 = [4,5,6,7]
print(l+l1) # merging
```

```python
l = [1,2,3,4]
print(l*3)
```

```python
# membership operator 
l = [1,2,3,4]
l1 = [4,5,6,[7,6]]
print(5 in l1)
print(5 not in l1)
print(7 in l1)
print([7,6] in l1)
```

```python
# loops
l = [1,2,3,4]
l1 = [4,5,6,7]
for i in l1: 
    print(i)
```

```python
l = [[[1,2,3,4,5],[6,7,8]]]
for i in l: 
    print(i)
```

```python
# list functions 
l = [1,2,3,4]
print(len(l))
print(min(l))
print(max(l))
print(sorted(l))
print(sorted(l,reverse=True))
```

```python
# count
l = [1,2,2,4,5,3,3,4]
print(l.count(2))
```

```python
# index 
l = [1,2,2,4,5,3,3,4]
l.index(2)
```

```python
# reverse
l = [1,2,2,4,5,3,3,4]
l.reverse()# reverse permanantly
print(l)
```

```python
# sort 
l = [1,2,2,4,5,3,3,4]
print(l)
print(sorted(l))
print(l)
l.sort() # sort permanantly
print(l)
```

```python
# copy 
l = [1,2,2,4,5,3,3,4]
print(id(l))
l1 = l.copy # makes a shallow copy( make a different address list )
print(l1)
print(id(l1))
```

```python
# list coprehension 
# provides a concise way of creating lists 
# add 1 to 10 numbers to a list 
l = []
for i in range(1,11): 
    l.append(i)
print(l)    
```

```python
l = [i for i in range(1,11)]
print(l)
```

```python
# scalar multiplication on a vector 
v = [2,3,4]
s = -3
l = [s*i for i in v]
print(l)
```

```python
# add squares 
l = [1,2,3,4,5]
[i**2 for i in l]
```

```python
# print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5 ==0]
```

```python
# find languages which start letter p
languages = ['java', 'python', 'php','c','javascript']
[i for i in languages if i.startswith('p')]
```

```python
# nested if with list comprehension 
basket = ['apple', 'guava', 'cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
[i for i in my_fruits if i in basket if i.startswith('a')]
```

```python
# 3*3 matrix using list comprehension 
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> list comprehension on 2 lists together 
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[i*j for i in l1 for j in l2]
```

```python
# 2 ways to traverse a list 
# 1. itemwise 2. indexwise
```

```python
# itemwise 
l = [1,3,2,4]
for i in l:
    print(i)
```

```python
# indexwise
l = [1,3,2,4]
for i in range(0,len(l)):
    print(i)
    print("list",l[i])
```

```python
# zip 
# write a program to add items of 2 lists indexwise 
l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]
list(zip(l1,l2))
[i+j for i,j in zip(l1,l2)]
```
