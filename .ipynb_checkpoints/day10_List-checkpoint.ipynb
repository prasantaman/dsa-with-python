```python
# list is flexible
# arrays are **homogeneous** (only one type of data type can store)whereas list is <u>hetrogeneous</u> 
# but execution time of array is lesser than list 
# occupi much space by list than array 
# list are stored in the form of referential array 
# list is a dynamic array
# characteristic of list-(orderes, changeble/mutable,can have duplicates,can be nested,accessible,can contain any type of object in python)
```

```python
a = 2 
print(id(2))
```

```python
l = [1,2,3]
print(id(l))
print(id(l[0]))
print(id(l[1]))
print(id(l[2]))
print(id(0))
print(id(1))
print(id(2))
```

```python
# empty
print([])
# 1D
print([1,2,3])
# 2d
print([[1,2,3],6,[4,5]])# hetrogeneous
# 3d
print([[[1,[3,2]],[2,3],[4,5],6]])
#hetrogeneous
print([1,True,5.6,5+6j,'Hello'])
# through type conversion
print(list('hello'))
```

```python
# indexing
l = [1,2,3,4,5]
print(l[0])
print(l[-1])
```

```python
l = [1,2,3,[4,5]]
print(l[-1])
print(l[-1][-2])
print(l[3][0])
```

```python
l = [[[1,[3,2]],[2,3],[4,5],6]]
print(l[0][0][1][0])
```

# adding items to a list 

```python
# append
l.append(True) # add items at the end
print(l)
```

```python
# extend(used to add multiple items at the end)
l.extend([6,7,8])
print(l)
```

```python
l.append([4,3,3])
print(l)
```

```python
l.extend('hello')
print(l)
```

```python
# insert
l = [1,2,3]
l.insert(1,True)
print(l)
```

# Editing items in a list 

```python
l = [1,2,3]
l[-1]= 100
print(l)
# editing with indexing
l[1:2] = [300,400]
print(l)
```
