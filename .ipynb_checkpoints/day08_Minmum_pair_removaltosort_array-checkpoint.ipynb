```python
#non decreasing checking 
ops = 0
def non_decreasing(nums): 
    for i in range(1,len(nums)): 
        if nums[i]< nums[i-1]:
            return False
    return True
while not non_decreasing(nums):
    min_sum = float('inf')
    idx = 0 
    for i in range(len(nums)-1):
        s = nums[i] +nums[i+1]
        if s<min_sum:
            min_sum = s
            idx = i
    nums = nums[:idx] + [nums[idx] +nums[idx+1]]+nums[idx+2:] 
    ops+=1
return ops
    
#non_decreasing([5,6,7,3,2,1])
non_decreasing([4,5,6])
```

```python
def minimumPairRemoval(nums):
    ops = 0
    def non_decreasing(arr): 
        for i in range(1,len(arr)):
            if arr[i]<arr[i-1]:
                return False
        return True
    while not non_decreasing(nums): 
        min_sum = float('inf')
        idx = 0 
        for i in range(len(nums)-1):
            s = nums[i]+nums[i+1]
            if s<min_sum:
                min_sum = s
                idx = i
        nums = nums[:idx]+[nums[idx]+nums[idx+1]]+nums[idx+2:]
        ops += 1
    return ops    
```

```python
print(minimumPairRemoval([1,2,5,2,3]))
```
