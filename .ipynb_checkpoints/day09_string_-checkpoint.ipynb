# Loops

```python
curr_pop = 10000
for i in range(10,0,-1): 
    print(i,curr_pop)
    curr_pop = curr_pop/1.1
```

```python
# Sequence Sum (1/1!+2/2!+3/3!-------)
n = int(input("enter a number"))
fact = 1
result = 0
for i in range(1,n+1):
    fact = fact*i
    result = result+ i/fact
print(result)    
```

```python
# Nested Loop 
rows = int(input("inter a number"))
for i in range(1,n+1): 
    for j in range(1,i+1):
        print(j,end = "")
    for k in range(i-1,0,-1):
        print(k, end = "")
    print()    
        
```

```python
lower = int(input("enter a number"))
upper = int(input("enter a number"))
for i in range(lower,upper+1): 
    for j  in range(2,i):
        if i%j == 0:
            break
    else: 
            print(i)
        
```

```python
# count the frequency of a particular character in a provided string 
s = input("enter an string")
x = input('what would u like to search for')
counter = 0 
for i in s:
    if i == x:
        counter +=1
print(counter)        
    
```

```python
# write a program which can remove a particular character from a string
x = input("enter an string")
y = input("what would like to remove")
result = ""
for i in x:
    if i!=y:
        result+=i
print(result)        
```

```python
# palindrome
x = input("enter an string")
flag = True
for i in range(0,len(x)//2):
    if x[i] != x[len(x)-i-1]:
        flag = False
        print('Not a prlindrome')
        break
if flag:
    print('palindrome')
        
```

```python
# write a program to count the number of words in a string without split()
x = input("enter an string")
temp = ''
l = []
for i in x:
    if i != ' ':
        temp +=i
    else:
        l.append(temp)
        temp = ' '
l.append(temp)
print(l)
    
```

```python
# convert a string to title case without using the title()
x = input("enter an string")
l = []
for i in x.split():
    l.append(i[0].upper() + i[1:].lower())
print(" ".join(l))    
```

```python
# write a program to convert an integer into string
number = int(input('enter a number:'))
digits = '0123456789'
result = ''
while number != 0:
    result = digits[number%10]+ result
    number = number//10
print(result)
print(type(result))
```
