```python
def is_even(num):
    """ even number calculation"""
    if type(num) == int:
        if num % 2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return f"Pagal hai kya {type(num)} likh raha hai"
```

```python
for i in range(0,11):
    x = is_even(i)
    print(x)
```

# *args and **kwargs 

```python
# special python keyword that are used to pass the variable length of arguments to a function
```

```python
# *args allows us to pass a variable number of non_keyword arguments to a function 
def multiply(*args):# here istead of args we can use any keyword like *puppy etc 
    product = 1
    for i in args:
        product = product * i
    print(args)    
    return product    
```

```python
multiply(2,3,4,5,6)
```

```python
# *kwargs allows us to pass any number of keyword arguments.
# keyword arguments mean that they contain a key-value pair, like a python dictonary
# orders of the arguments matter(normal->*args->**kwargs)
def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,"->", value)
```

```python
display(india = 'delhi', pak = "colony")
```

# Nested function 

```python
def f():
    def g():
        print("inside function g")
    g()    
    print("inside function f") 
f()
```

```python
g()# ander vala function hidden hota hai. ander vala function baher vala function hi access ker sakta hai 
```

```python
def f(): # infinite loop 
    def g():
        print("inside function g")
        f()
    g()    
    print("inside function f") 
f()
```

```python
def g(x):
    def h():
        x = 'abc'
    x = x+1
    print('in g(x): x =',x)
    h()
    return x
x = 3
z = g(x)
```

```python
def g(x):
    def h(x):
        x = x+1
        print('in h(x): x=',x)
    x = x+1
    print('in g(x): x=', x)
    h(x)
    return x
x = 3
x = g(x)
```

# lambda function 
##### is a small anonymous function 
##### a lambda function can take any number of arguments, but can only have one expression


```python
a = lambda x:x**2
a(2)
```

```python
a = lambda x,y: x+y
a(3,4)
```

# diff between lambda vs normal function 
- no name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable
##### they are used with HOF(higher order function)

```python
a = lambda s:'a' in s
a('hello')
```

```python
a = lambda x:'even' if x%2==0 else 'odd'
a(5)
```

# higher order function 

```python
def square(x): 
    return x**2
# hof
def transform(f,l): 
    output = []
    for i in l:
        output.append(f(i))
    print(output)    
l = [1,2,3,4,5]
transform(lambda x:x**3,l)
```

# map

```python
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
l = [1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd', l))
```

# filter

```python
l = [3,4,5,6,7,8]
list(filter(lambda x:x>5,l))
```

```python
fruits = ['apple','guava','cherry']
list(filter(lambda x:x.startswith('a'),fruits))
```

# reduce

```python
import functools
functools.reduce(lambda x,y: x+y,[1,2,3,4,5])
```

```python
functools.reduce(lambda x,y:x if x<y else y,[23,34,545,3,1,0])
```
