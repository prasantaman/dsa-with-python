# tuple functions (len,sum, min, max, sorted,count,index)
# diff between tuple and list 
- Syntax
- Mutability(list are mutable)
- Speed (tuples are faster than list)
- Memory(tuple takes less space)
- built in functinality
- error prone

```python
# Speed check
import time 
l = list(range(100000))
t = tuple(range(100000))
start = time.time()
for i in l:
    i*5
print('list time:', time.time()-start)

ab = time.time()
for i in t:
    i*5
print('tuple time:', time.time()-ab)    
```

```python
# size check
import sys 
l = list(range(1000))
t = tuple(range(1000))
print('list size:', sys.getsizeof(l))
print('tuple size:', sys.getsizeof(t))
```

# set operations 

```python
# union 
a = {1,2,3,4,5,6}
b = {4,5,6,7,7}
a | b
```

```python
# intersection
a = {1,2,3,4,5,6}
b = {4,5,6,7,7}
print(a&b)
# difference 
print(a-b)
print(b-a)
# symmetric difference 
print(a^b)
# membership operator
print(1 not in a)
# iteration 
for i in a:
    print(i)
```

# functions 

```python
# union/ update 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,7}
s1.union(s1)
print(s1)
s1.update(s2)
print(s1)
print(s2)
```

```python
# intersection/intersection_update 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,7}
s1.intersection(s1)
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference/difference_update 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,7}
s1.difference(s1)
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,7}
s1.symmetric_difference(s1)
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# disjoint 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,7}
s1.isdisjoint(s2) # false cuz there exist common values 
```

```python
# issubset 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6}
print(s1.issubset(s2))
print(s2.issubset(s2))
```

```python
# issuperset 
s1 = {1,2,3,4,5,6}
s2 = {4,5,6}
print(s1.issuperset(s2))
print(s2.issuperset(s1))
```

# dictionary

```python
# also known as map or associative arrays
# mutable 
# indexing has no meaning 
# keys can't be duplicated 
# keys can't be mutable items 
```

```python
# empty dict 
a = {}
print(a)
# mixed dict 
b = {(1,2,3): 1, 'hello': 'world'}
print(b)
# 2d dict 
c = {
    'name': "prasant",
    'sem': 7,
    'subject':{
        'dsa':54,
        'math': 78
    },
    'class' : 10
}
print(c)
# using sequence and dict function 
d = dict([(1,1),('hello','world'),(3,2)])
print(d)
# duplicate keys 
e= dict([(1,1),('hello','world'),(1,2)])
print(e)
# mutable items as keys
f = {'name':'prasant',(1,2,3):4}
print(f)
```

```python
g = {'name':'prasant',[1,2,3]:4}
print(g)
```

```python
c = {
    'name': "prasant",
    'sem': 7,
    'subject':{
        'dsa':54,
        'math': 78
    },
    'class' : 10
}
print(c['name'])
print(c.get('subject'))
```

```python
# adding key_value pair 
c['gender'] = 'male'
c
```

```python
# remove key value pair 
c = {
    'name': "prasant",
    'sem': 7,
    'subject':{
        'dsa':54,
        'math': 78
    },
    'class' : 10
}
# pop 
c.pop('class')
print(c)
# popitem
c.popitem()# remove last item
print(c)
# del
del c['name']
print(c)
# clear 
c.clear()
print(c)
```

```python
# dictionary comprehension 
# print first numbers and their squares 
{i:i**2 for i in range(1,11)}
```

```python
distances = {'delhi': 1000, 'mumbai':2000,'bangalore':3000}
{key:value*0.62 for(key,value) in distances.items()}
```

```python
# using zip 
days = ['sun','mon','tue','thu','fri','sat','sun']
temp_c = [30.5,23,64,34,23,54,64]
{i:j for i,j in zip(days,temp_c)}
```

```python
# using if condition
products = {'phone':10,"lap": 0,'char':32}
{key:value for (key,value) in products.items() if value>0}
```

```python
# nested comprehension 
{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
