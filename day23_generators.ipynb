```python
def num(x):
    for i in range(1, x+1):
        yield i

y = num(10)
```

```python
print(next(y))
print(next(y))
```

```python
for i in y:
    print(i)
```

# Range function using generators

```python
def my_range(start,end):
    for i in range(start,end):
        yield i
        
```

```python
for i in my_range(15,26):
    print(i)
```

# generator Expression

```python
L = [i**2 for i in range(1,101)]
```

```python
print(L)
```

```python
gen = (i**2 for i in range(1,101))
for i in gen:
    print(i)
```

# why generators

```python
#memory efficient
l = [x for x in range(10000)]
gen = (x for x in range(10000))
import sys 
print('size of l in memory',sys.getsizeof(l))
print('size of gen in memory', sys.getsizeof(gen))
```

```python
# representation infinite streams
def all_even():
    n = 0
    while True:
        yield n
        n+=2
```

```python
even_num_gen = all_even()
next(even_num_gen)
next(even_num_gen)
next(even_num_gen)
next(even_num_gen)
```
