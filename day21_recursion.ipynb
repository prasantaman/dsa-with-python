```python
def fact(num):
    if num ==1:
        return 1
    else:
        return num *fact(num-1)

fact(5)
```

```python
def palin(text):
    if len(text) ==1:
        print("palindrome")
    else: 
        if text[0] == text[-1]:
            palin(text[1:-1])
        else:
            print('not a palin')
palin('mam')
```

```python
# the rabbit problem 
import time
def fib(x):# but it's slow 
    if x==0 or x==1:
        return 1
    else:
        return fib(x-1) +fib(x-2)

start = time.time()

print(fib(5))
print(time.time() - start)
```

```python
def memo(m,d):
    if m in d:
        return d[m]
    else:    
        d[m] = memo(m-1,d) + memo(m-2,d)
        return d[m]

start = time.time()
d = {0:1,1:1}
print(memo(100,d))
print(time.time() - start)
```
